```python
import numpy as np
import matplotlib.pyplot as plt


category_names = ['rest and leisure', 'medical treatment',
                  'visit to relatives', 'studies', 'business', 'commerce']
results = {
    '2012': [9.6, 4.1, 80.7, 0.3, 4.3, 1.0],
    '2013': [7.9, 4.7, 80.3, 0.4, 5.3, 1.4],
    '2014': [8.4, 4.6, 79.2, 0.5, 5.8, 1.5],
    '2015': [8.7, 2.9, 81.1, 0.6, 5.9, 0.8],
    '2016': [8.7, 1.3, 81.5, 0.3, 7.4, 0.8],
    '2017': [7.6, 1.1, 84.5, 0.2, 5.9, 0.7],
    '2018': [8.6, 1.0, 88.2, 0.3, 1.0, 0.9],
    '2019': [15.5, 0.8, 81.8, 0.3, 0.8, 0.8],
    '2020': [8.6, 1.0, 87.9, 0.3, 1.2, 1.1],
    '2021': [8.2, 1.7, 85.7, 0.3, 3.2, 0.7],
}

def survey(results, category_names):
   
    labels = list(results.keys())
    data = np.array(list(results.values()))
    data_cum = data.cumsum(axis=1)
    category_colors = plt.colormaps['RdYlGn'](
        np.linspace(0.15, 0.85, data.shape[1]))

    fig, ax = plt.subplots(figsize=(9.2, 6))
    ax.invert_yaxis()
    ax.xaxis.set_visible(False)
    ax.set_xlim(0, np.sum(data, axis=1).max())

    for i, (colname, color) in enumerate(zip(category_names, category_colors)):
        widths = data[:, i]
        starts = data_cum[:, i] - widths
        rects = ax.barh(labels, widths, left=starts, height=0.8,
                        label=colname, color=color)

        r, g, b, _ = color
        text_color = '0' if r * g * b < 0.9 else 'darkgrey'
        
    ax.legend(ncol=len(category_names), bbox_to_anchor=(0, 0),
              loc='lower left', fontsize='small')
    return fig, ax
        
survey(results, category_names)

plt.xticks(rotation = 0)
plt.title("Arrivals of foreign citizents to the Republic of Uzbekistan, %")
plt.ylabel("YEARS")

plt.show()
```
